```python
# Data Normalization
# Adventages: Similar value Range + Similar intrinsic
# influence on analytical model.
# 3 technics to normalize data
# Simple Feature Scaling: Xnew= Xold/Xmax
df[name_variable] = df[name variable]/df[name variable].max()
# Min-Max: (Xold-Xmin)/(Xmax-Xmin)
df[name_variable] = (df[name variable]-df[name variable].min())/ \
                    (df[name variable].max()-df[name variable].min())
# Z-score: (Xold-median)/standard deviation
df[name_variable] = (df[name variable]-df[name variable].mean())/ \
                      df[name variable].max()
```
