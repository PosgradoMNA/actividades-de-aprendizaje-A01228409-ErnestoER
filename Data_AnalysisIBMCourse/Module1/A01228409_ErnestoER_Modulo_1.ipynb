# Importing and Exporting Data in Python

```python
# Datasets format .csv, .json, .xlsx, .hdf
ds_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'

# Importing a CSV into python
# Import pandas library for importing library
import pandas as pd

df = pd.read_csv(ds_url) # without header header=None
df.head()
df.tail()
# .head(n) show the first n rows of dataframe (default 5)
# .tail(n) show bottom n rows
# Reclace headers name df.columns = [headers_name]
# Data Format     Read             Save
# csv            pd.read_csv()     df.to_csv()
# json           pd.read_json()    df.to_json()
# Excel          pd.read_excel()   df.to_excel()
# Sql            pd.read_sql()     df.to_sql()
```
