```python
# Path for testing write text
w_file_path = "/Drive/MyDrive/Ciencia y analitica de datos/IBM Data Science Course/write_file.txt"
# File modes. 
# 'a' append -> write but not allow to create a file
# 'w' only write and allow to create a new file
# 'r' only read 
with open(w_file_path, "w") as WriteFile:
  WriteFile.write("This is line A\n")
  WriteFile.write("This is line B")

# Read what we wrote on txt
with open(w_file_path, "r") as ReadFile:
  file_content = ReadFile.read()
  print(file_content)
```

```python
# Path for testing read and write text
read_file_path = "/Drive/MyDrive/Ciencia y analitica de datos/IBM Data Science Course/read_file.txt"
r_n_w_file_path = "/Drive/MyDrive/Ciencia y analitica de datos/IBM Data Science Course/read_n_write_file.txt"
with open(read_file_path, "r") as readfile:
  with open(r_n_w_file_path, "w") as writefile:
    for line in readfile:
      writefile.write(line)

with open(r_n_w_file_path, "r") as new_file:
  file_content = new_file.read()
  print(file_content)
```

```python
import pandas as pd

# Create Data frame from dictionary
songs = {'Album': ['Thriller', 'Back in Black', 'The Dark Side of the moon',\
'The Bodyguard', 'Bat Out of Hell'],
'Released': [1982, 1980, 1973, 1992, 1977],
'Length': ['00:42:19', '00:42:11', '00:42:49', '00:57:44', '00:46:33']}
songs_frame = pd.DataFrame(songs)
print(songs_frame.head())

# Get a specific column of data frame
split_data_frame = songs_frame[['Album']]
print(split_data_frame)

# Access unique data
# ix is not working anymore
print(songs_frame.iloc[1,0])
print(songs_frame.loc[0, 'Album'])
```

```python
# Working with and Saving Data with Pandas

songs = {'Album': ['Thriller', 'Back in Black', 'The Dark Side of the moon',\
'The Bodyguard', 'Bat Out of Hell'],
'Released': [1982, 1980, 1973, 1992, 1977],
'Length': ['00:42:19', '00:42:11', '00:42:49', '00:57:44', '00:46:33']}
songs_frame = pd.DataFrame(songs)

# Save only songs released after 1980
modified_songs_frame = songs_frame[songs_frame['Released']>=1980]

save_csv_path = "/Drive/MyDrive/Ciencia y analitica de datos/IBM Data Science Course/songs_released_after_1980.csv"
# Save csv
modified_songs_frame.to_csv(save_csv_path)
```
